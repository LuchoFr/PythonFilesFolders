```python
from pathlib import Path
import os 
import shutil

base=Path.home()
direccion_recetas=Path(base,"Recetas") #estoy en \Recetas 
diccionario_categorias = {}#va a ser mi diccionario con las categorias (carnes, ensaladas, postre, pastas y mas si llego a agregar o menos si elimino) y con sus direcciones en el valor
for carpeta in direccion_recetas.glob('*'): #cargo la lista con la tupla de (categoria, y direccion)
    if carpeta.is_dir():
        diccionario_categorias[carpeta.name]=carpeta
print(diccionario_categorias)

def verificador_categorias(categoria):
    if categoria in diccionario_categorias:
        return True
    else:
        return False
    
def verificador_recetas(ruta_receta):
        # Verifica si el archivo existe
    if os.path.isfile(ruta_receta):
        return True
    else:
        return False 

def elegir_categoria():
    print("Elja una categoria: ")
    for claves in diccionario_categorias.keys():
        print(f"- {claves} ")
    seleccion=input("Su eleccion: ")
    return seleccion


def elegir_receta(categoria):
     print("Elija una receta: ")
     for txt in Path(diccionario_categorias[categoria]).glob("*.txt"):
          print(f"-{txt.stem}")
     seleccion=input("Su elección: ")
     return seleccion

def leer_receta(archivo):
    mi_archivo=open(archivo,mode="r")
    receta=mi_archivo.read()
    mi_archivo.close()
    return receta

def armador_de_ruta_receta(ruta_categoria,receta): #recibe la ruta de la categoria y la receta le agrega el txt y transforma a la direccion absoluta de la receta 
    receta_txt=receta+".txt"
    ruta=Path(ruta_categoria,receta_txt)
    return ruta 


def crear_receta(categoria,nombre_receta,texto): #recibe una categoria el nombre que le qremos poner a la receta y el texto de lo que queremos poner
    ruta_categoria=diccionario_categorias[categoria]
    nombre_receta_txt=nombre_receta+".txt"
    ruta_nueva_receta=Path(ruta_categoria,nombre_receta_txt)
    ruta_nueva_receta.touch()
    with open(ruta_nueva_receta, "w") as archivo: #otra forma de abrir con el with q te no tenes q poner open write close
        archivo.write(texto)

def eliminar_receta(categoria,nombre_receta):
    ruta_categoria=diccionario_categorias[categoria]
    ruta_eliminar_receta=armador_de_ruta_receta(ruta_categoria,nombre_receta)
    os.remove(ruta_eliminar_receta)

def eliminar_categoria(nombre_categoria):
    ruta_eliminar_categoria=Path(direccion_recetas,nombre_categoria)
    shutil.rmtree(ruta_eliminar_categoria) # Elimina el directorio y su contenido


def crear_categoria(nombre_categoria):
    ruta_nueva_categoria=Path(direccion_recetas,nombre_categoria)
    ruta_nueva_categoria.mkdir()


r=armador_de_ruta_receta(diccionario_categorias["Carnes"],"Matambre a la Pizza")
print(leer_receta(r))


```

```python
for i in range(0,10):
    print("\n")
    
bandera=True

while bandera:
    print("Bienvenido al recetario: que opcion desea elegir: \n [1] - Leer receta \n [2] - Crear receta \n [3] - Crear categoría \n [4] - Eliminar receta \n [5] - Eliminar categoría \n [6] - Finalizar programa\n")
    opcion=input("Su elección: ")
    if opcion == "1":
        cat=elegir_categoria()
        print("")
        if verificador_categorias(cat):
            rec=elegir_receta(cat)
            ruta_receta=armador_de_ruta_receta(diccionario_categorias[cat],rec)
            print("")
            if verificador_recetas(ruta_receta):
                print("")
                print(leer_receta(ruta_receta))
                print("")
            else:
                print("La receta no existe \n")
        else:
            print("La categoria no existe \n")
            
    elif opcion == "2":
        cat=elegir_categoria()
        nombre_nueva_receta=input("Ingrese el nombre de la receta que quiere crear: ")
        ruta_receta=armador_de_ruta_receta(diccionario_categorias[cat],nombre_nueva_receta)
        if not verificador_recetas(ruta_receta):
            texto_receta=input("Ingrese las instrucciones de la receta: ")
            crear_receta(cat,nombre_nueva_receta,texto_receta)
            print("")
            print(f"La receta {nombre_nueva_receta} fue cargada con exito \n")
        else:
            print("")
            print("La receta que desea ingresar ya existe \n")

    elif opcion == "3":
        cat=input("Ingrese el nombre de la nueva categoría que desea crear: ")
        if not verificador_categorias(cat):
            crear_categoria(cat)
            print("")
            print(f"La categoría {cat} fue creada con exito \n")
            diccionario_categorias[cat]=Path(direccion_recetas,cat)
        else:
            print("")
            print("Esa categoria ya existe \n")

    elif opcion == "4":
        cat=elegir_categoria()
        print("")
        rec=elegir_receta(cat)
        ruta_receta=armador_de_ruta_receta(diccionario_categorias[cat],rec)
        if verificador_recetas(ruta_receta):
            eliminar_receta(cat,rec)
            print("")
            print(f"La receta {rec} fue eliminada exitosamente \n")
        else:
            print("")
            print("La receta que desea eliminar no existe \n")

    elif opcion == "5":
        print("Recuerde que si elimina la categoría elimina todas las recetas que contiene dentro \n")
        cat=elegir_categoria()
        print("")
        if verificador_categorias(cat):
            eliminar_categoria(cat)
            print(f"La categoria {cat} fue eliminada con exito \n")
            del diccionario_categorias[cat]
        else:
            print("")
            print("La categoria que desea eliminar no existe\n")
            
    elif opcion == "6":
        bandera=False
    else:
        print("Por favor ingrese una opción valida \n")



        


```
